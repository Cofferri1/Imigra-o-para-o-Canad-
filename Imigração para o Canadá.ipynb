```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Carregar os dados
arquivo = '/content/imigrantes_canada.csv'
df = pd.read_csv(arquivo)
df.head()
```

```python
#Verificar o type de cada coluna
df.info()
```

```python
df.set_index('País', inplace=True)
```

```python
anos = list(map(str, range(1980, 2014)))
```

```python
#Ano #Numero de imigrante
brasil = df.loc['Brasil', anos]
```

```python
#Criar um dict para verificar a quantidade de imigrantes do Brasil
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
#Criar uma grafico referente ao ano e aos imigrantes
fig, ax = plt.subplots(figsize=(8, 4))
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'], lw = 3, color = 'green')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010', '2014'])
plt.title('Imigrantes no Brasil')
plt.xlabel('Ano')
plt.ylabel('Imigrantes')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

fig.savefig('Imigrantes do Brasil para o Canadá.png', transparent=False, dpi=300, bbox_inches='tight')
plt.show()
```

```python
#criar duas figuras comparativas
fig, ax = plt.subplots(1,2 ,figsize=(15, 5))
ax0 = ax[0]
ax1 = ax[1]
ax0.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
ax0.set_title('Imigrantes no Brasil')
ax0.set_xlabel('Ano')
ax0.set_ylabel('Número de Imigrantes')
ax0.xaxis.set_major_locator(plt.MultipleLocator(5))
ax0.grid()

ax1.boxplot(dados_brasil['imigrantes'])
ax1.set_title('Imigrantes no Brasil')
ax1.set_xlabel('Ano')
ax1.set_ylabel('Número de Imigrantes')
ax1.grid()
```

```python
#Varias figuras

fig, axs = plt.subplots(2, 2, figsize=(10, 6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Imigração dos Países para Canadá')

axs[0, 0].plot(df.loc['Brasil', anos], color ='g')
axs[0, 0].set_title('Brasil')

axs[0,1].plot(df.loc['Colômbia', anos])
axs[0,1].set_title('Colômbia')

axs[1,0].plot(df.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Peru', anos])
axs[1,1].set_title('Peru')

#Percorer todos os axs para alinhar a legenda do X
#Corrigir o valor Y com minimo e máximo
ymin = 0
ymax = 8000
for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.set(xlabel='Ano', ylabel='Número de Imigrantes')
  ax.grid()
  ax.set_ylim(ymin, ymax)

plt.show()
```

```python
lojas = ['A', 'B', 'C', 'D']

vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}

df = pd.DataFrame(vendas_2022, index=lojas)

# Criar a figura e os subplots
fig, axs = plt.subplots(2, 2, figsize=(14, 8))

# Ajustar os espaçamentos entre os subplots
plt.subplots_adjust(wspace=0.35, hspace=0.35)

# Adicionando um título geral para os subplots
fig.suptitle('Vendas no período de janeiro a dezembro de 2022 nas lojas A,B,C e D', fontsize=18)

# Cores para cada loja
cores = ['blue', 'green', 'red', 'purple']

# Adicionar os gráficos em cada um dos subplots com melhorias
for i, loja in enumerate(lojas):
      row = i // 2  # Calcula a linha do subplot
      col = i % 2   # Calcula a coluna do subplot
      axs[row, col].plot(df.loc[loja], color=cores[i], marker='o', linestyle='-', lw=1)  # Adiciona marcadores e estilo de linha
      axs[row, col].set_title(f'Vendas na loja {loja}', fontsize=14)  # Aumenta o tamanho da fonte do título
      axs[row, col].set_xlabel('Mês', fontsize=12)  # Adiciona rótulo do eixo x com tamanho de fonte
      axs[row, col].set_ylabel('Número de vendas', fontsize=12)  # Adiciona rótulo do eixo y com tamanho de fonte
      axs[row, col].grid(linestyle= '--')  # Adiciona grid
      axs[row, col].tick_params(axis='x', rotation=45)  # Rotaciona os rótulos do eixo x para melhor legibilidade

# Exibir a figura
plt.show()
```

```python
lojas = ['A', 'B', 'C', 'D']

vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}

df = pd.DataFrame(vendas_2022, index=lojas)

plt.subplots(2,2, figsize=(10, 6))
plt.subplots_adjust(wspace=0.35, hspace=0.45)
plt.suptitle('Vendas no período de janeiro a dezembro de 2022 nas lojas A,B,C e D', fontsize=18)
cores = ['blue', 'green', 'red', 'purple']

for i, loja in enumerate(lojas):
  plt.subplot(2, 2, i+1)
  plt.plot(df.loc[loja], marker='o', linestyle='-', lw=1, color=cores[i])
  plt.title(f'Vendas na loja {loja}', fontsize=14)
  plt.xlabel('Mês', fontsize=12)
  plt.ylabel('Número de vendas', fontsize=12)
  plt.tick_params(axis='x', rotation=45)
  plt.grid(linestyle= '--')
  plt.tight_layout()

plt.show()
```

```python
america_sul = df.query('Região == "América do Sul"')
america_sul
cores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown', 'slateblue', 'gray', 'olive', 'navy', 'teal', 'tomato']

fig, ax = plt.subplots(figsize=(12, 5))
ax.barh(america_sul['País'], america_sul['Total'], color=cores)
ax.set_title('Imigração da América do Sul para o Canadá\n1980 a 2013', fontsize=18)
ax.set_ylabel('Número de imigrantes', fontsize=14)
ax.set_xlabel('País', fontsize=14)
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_tick_params(labelsize=12)

plt.show()
```

```python
america_sul = df.query('Região == "América do Sul"')
america_sorted = america_sul.sort_values('Total', ascending=True)
cores = []

for pais in america_sorted['País']:
  if pais == 'Brasil':
    cores.append('green')
  else:
    cores.append('grey')

fig, ax = plt.subplots(figsize=(12, 5))
ax.barh(america_sorted['País'], america_sorted['Total'], color=cores)
ax.set_title('Imigração da América do Sul para o Canadá\n1980 a 2013', fontsize=18)
ax.set_ylabel('Número de imigrantes', fontsize=14)
ax.set_xlabel('País', fontsize=14)
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_tick_params(labelsize=12)

#interar o valor para cada item

for i, v in enumerate(america_sorted['Total']):
  ax.text(v + 20, i, str(v), color='black', fontsize=10, ha='left', va = 'center')

ax.set_frame_on(False)

plt.show()
```
